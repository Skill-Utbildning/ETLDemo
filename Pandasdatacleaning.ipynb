But now we want only records with a contraband found.

```python
contraband = result.loc[result['contraband_found'].isnull()==False]
```

```python
contraband.head()
```

```python
contraband.isnull().sum()
```

Lets remove some coded data that just encode text fields. Warning, these are the type of fields we would like to keep or make by encoding for ML or statistics.

```python
cleaned_contra = contraband.drop(['raw_DriverRace', 'raw_ReasonForStop', 'raw_TypeOfMovingViolation','raw_ResultOfStop'], axis = 'columns')
cleaned_contra.shape
```

```python
cleaned_contra.head()
```

There is subject_age field thats not empty, worth keeping?

```python
aged = cleaned_contra[cleaned_contra.subject_age.notna()]
aged
```

That was not very exiting. Age is not very prevalent. We can actually see all.
